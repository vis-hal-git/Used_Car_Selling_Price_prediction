```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("car.csv")
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data["Mileage"]=data["Mileage"].str.split(' ').str.get(0)
data["Engine"]=data["Engine"].str.split(' ').str.get(0)
data["Power"]=data["Power"].str.split(' ').str.get(0)
```

```python
data.head()
```

# car_name

```python
car_Name_le=LabelEncoder()
data["Car_Name"]=car_Name_le.fit_transform(data["Car_Name"])
```

# fuel_type

```python
data["Fuel_Type"].unique()
data["Transmission"].unique()
```

```python
Fuel_Type_le=LabelEncoder()
data["Fuel_Type"]=Fuel_Type_le.fit_transform(data["Fuel_Type"])
data["Fuel_Type"]
```

# transmission

```python
Transmission_le=LabelEncoder()
data["Transmission"]=Transmission_le.fit_transform(data["Transmission"])
```

# Owner_Type

```python
Owner_Type_le=LabelEncoder()
data["Owner_Type"]=Owner_Type_le.fit_transform(data["Owner_Type"])
```

# Mileage

```python
Mileage_le=LabelEncoder()
data["Mileage"]=Mileage_le.fit_transform(data["Mileage"])
```

# Engine

```python
Engine_le=LabelEncoder()
data["Engine"]=Engine_le.fit_transform(data["Engine"])
```

# Power

```python
Power_le=LabelEncoder()
data["Power"]=Power_le.fit_transform(data["Power"])
```

```python
data.head()
```

```python
Input_data=data.iloc[:,:-1]
Output_data=data["Selling_Price"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data.columns
```

```python
sp=StandardScaler()
Input_data=pd.DataFrame(sp.fit_transform(Input_data),columns=Input_data.columns)

```

```python
Input_data
```

```python
sns.heatmap(data=data.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(Input_data,Output_data,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr=DecisionTreeRegressor()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
sv=SVR()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100
```

```python
lr=KNeighborsRegressor(n_neighbors=300)
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)
RFR.score(x_test,y_test)*100, RFR.score(x_train,y_train)*100
```
